#### Ejemplo para cargar un archivo a un dataframe en Pandas

##### - Se va a utilizar al función "read_csv" <br>- Por defecto, pandas toma el separador "," (coma), como este archivo está separado por ";" (punto y coma) se debe especificar con el parámetro "sep"

```python
import pandas as pd
pd.read_csv('dataset_bank.csv', sep=";")
```
